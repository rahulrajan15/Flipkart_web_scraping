```python
### To iterate the different flipkart pages using the url and beautifulsoup liberaries
```

```python
for i in range(2,10):
    url = "https://www.flipkart.com/search?q=mobile%20phone%20under%2050000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"lxml")
    np = soup.find("a",class_ = "_1LKTO3").get("href")
    cnp = "https://www.flipkart.com"+np
    print(cnp)
```

```python
### Scarping data from single page & create empty list to store the data
```

```python
Pnames = []
Prices = []
Description = []
Ratings = []

for i in range(2,12):
    url = "https://www.flipkart.com/search?q=mobile%20phone%20under%2040000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"lxml")

    box = soup.find("div","_1YokD2 _3Mn1Gg")

    # iterate phone names from page 1
    pnames = box.find_all("div",class_ = "_4rR01T")
    # we are using for loop for eleminating div tag text in the result
    for i in pnames:
        name = i.text
        Pnames.append(name)   
    #print(Pnames)
    #print("There total of",len(Pnames), "mobiles in page 1")


    # iterate phone prices from page 1
    prices = box.find_all("div",class_ = "_30jeq3 _1_WHN1")
    # we are using for loop for eleminating div tag text in the result
    for i in prices:
        price = i.text
        Prices.append(price)  
    #print(Prices)
    #print("There total of",len(Prices), "mobiles & its prices in page 1")

    # iterate phone description from page 1
    description = box.find_all("ul",class_ = "_1xgFaf")
    # we are using for loop for eleminating div tag text in the result
    for i in description:
        desc = i.text
        Description.append(desc) 
    #print(Description)
    #print("There total of",len(Description), "mobiles & there descriptions in page 1")

    # iterate phone ratings from page 1, 
    # we need to select ratings of the page 1 & since its showing all html page ratings(ie; 34 instead of 24) so select the div tag of that particular 24 mobiles div tag only
    ratings = box.find_all("div",class_ = "_3LWZlK")
    # we are using for loop for eleminating div tag text in the result
    for i in ratings:
        rate = i.text
        Ratings.append(rate)    
    #print(Ratings)
    #print("There total of",len(Ratings), "mobiles & its real-time ratings in page 1")



# Create dataframe and convert all the collected data into csv file
df = pd.DataFrame({"Product Name":Pnames,"Product Price":Prices,"Description":Description,"Ratings":Ratings})
#print(df)

#In above print function we can see that all necessary data is displayed with their values

# Now convert the data into excel or csv file format & change the backward slashes into forward
df.to_csv("F:/Courses_Files/Data_Analyst_Files_&_Resources/python_DA_Projects/togithub_DA/Web_Scraping_projects/flipkart_mobiles_under_40000.csv")

```
